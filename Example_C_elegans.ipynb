```python
from dw2d import geometry_reconstruction_2d
import skimage.io as io
label = io.imread('data/c_elegans_label.png')
img = io.imread("data/c_elegans_image.png")
DW = geometry_reconstruction_2d(label,expansion_labels=0, min_dist=3,original_image=img)
DW.extended_plot()
```

```python
from foambryo2d import infer_tension,plot_tension_inference
Mesh = DW.return_dcel()
_,dict_tensions,_ = infer_tension(Mesh,mean_tension=1,check_topo_changes=False,mode='YD')
plot_tension_inference(Mesh,dict_tensions,original_image = img)
```
